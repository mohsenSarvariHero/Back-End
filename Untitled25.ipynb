```python
import matplotlib.pyplot as plt

# نقاط مورد بررسی
specificities = [1, 0.75, 0.5, 0.02, 0]
sensitivities = [0.24, 0.8, 0.9, 0.75, 1]

# تعداد بیماران و سالم‌ها
P = 145  # بیماران (واقعی مثبت)
N = 104  # سالم‌ها (واقعی منفی)

# محاسبه Accuracy برای هر نقطه
accuracies = []
for sens, spec in zip(sensitivities, specificities):
    TP = round(sens * P)
    FN = P - TP
    TN = round(spec * N)
    FP = N - TN
    accuracy = (TP + TN) / (TP + TN + FP + FN)
    accuracies.append(accuracy)

# رسم نمودار Accuracy
plt.figure(figsize=(7, 5))
plt.plot(specificities, accuracies, marker='o', linestyle='-', color='b', label="Accuracy")

# افزودن نقاط داده‌ها به نمودار
for x, y in zip(specificities, accuracies):
    plt.text(x, y, f"{y:.2f}", fontsize=10, verticalalignment='bottom', horizontalalignment='right')

plt.xlabel("1-Specificity")
plt.ylabel("Accuracy")
plt.title("Accuracy vs. Specificity")
plt.ylim(0, 1.1)
plt.grid(True)
plt.legend()
plt.show()
```
